```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DATA_DIR = "../gamedata/original_dataset/"
DATA_FILE = "dd_event_schemas.csv"
GenreStats_DIR = "../gamedata/genreStats/"
PlatformStats_DIR = "../gamedata/platformStats/"
GenreScore_DIR = "../gamedata/genreScore/"
PlatformScore_DIR = "../gamedata/platformScore/"
WeightedScore_DIR = "../gamedata/weightedScore/"
```

```python
data = pd.read_csv(DATA_DIR+DATA_FILE)
```

```python
genres = data.drop_duplicates(['genre'])
platforms = data.drop_duplicates(['platform'])
#genres = genres['genre']
```

```python
selectGenre = data[data['genre'] == 'No genre']
drop=selectGenre.drop_duplicates(['gameid','eventname'])
```

```python
for genre in genres['genre']:
    selectGenre = data[data['genre'] == genre]#Select records with a specified genre
    dataAfterDrop=selectGenre.drop_duplicates(['gameid','eventname'])#Drop duplicate data 
    count1 = dataAfterDrop['eventname'].value_counts(normalize=True)#normalize the results#Count frequency of event occurance
    #count2 = count1.drop_duplicates['eventname']
    count2 = dataAfterDrop.drop_duplicates(['eventname'])
    num = len(count2)#Number of events 
    print(genre)
    print('Number of events:','' , num)
    print(count1[:10])
    fig = plt.figure(figsize=(6,5))
    count1[:10].sort_values().plot(kind='barh',title=genre)
    #for a,b in zip(x,y):
    #    plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=7)
    plt.show();
    print("")
    #header1=['eventname','score']
    #header2=DataFrame(header)
    #header2.to_csv(DATA_DIR+genre+'.csv',sep=',', header=False, index=False)  
    #count1.to_csv(DATA_DIR+genre+'.csv', header=False)
    count1.to_csv(GenreStats_DIR+genre+'Stat.csv', header=False)
```

```python
print('Number of platforms:','' , len(platforms['platform']))#Num of platform in the dataset
for platform in platforms['platform']:
    selectPlatform = data[data['platform'] == platform]#Select records with a specified platform
    dataAfterDrop=selectPlatform.drop_duplicates(['gameid','platform'])#Drop duplicate eventname per game
    count1 = dataAfterDrop['eventname'].value_counts(normalize=True)#normalize the results#Count frequency of event occurance
    count2 = dataAfterDrop.drop_duplicates(['eventname'])
    num = len(count2)#Number of events 
    print(platform)
    print('Number of events:','' , num)
    print(count1[:30])
    fig = plt.figure(figsize=(6,5))
    count1[:10].sort_values().plot(kind='barh',title=platform)
    plt.show();
    print("") 
    #count1.to_csv(DATA_DIR+platform+'.csv', header=False)
    count1.to_csv(PlatformStats_DIR+platform+'Stat.csv', header=False)
```

```python
#Use the frequency in genre multiply 100 as the score of a eventname
for genre in genres['genre']:
    genreScore = pd.read_csv(GenreStats_DIR+genre+'Stat.csv',header=None)
    genreScore.iloc[:,1] = genreScore.iloc[:,1]*100
    genreScore.to_csv(GenreScore_DIR+genre+'Score.csv',header=False,index=False)
    print(genre)
    print(genreScore)
    print("")
    
```

```python
#Use the frequency in genre multiply 100 as the score of a eventname
for platform in platforms['platform']:
    platformScore = pd.read_csv(PlatformStats_DIR+platform+'Stat.csv',header=None)
    platformScore.iloc[:,1] = platformScore.iloc[:,1]*100
    platformScore.to_csv(PlatformScore_DIR+platform+'Score.csv',header=False,index=False)
    print(platform)
    print(platformScore[:10])
    print("")
```

```python
def weightedScore(genreScore,platformScore):
    return genreScore.iloc[:,1]*platformScore[:,1]
```
