```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
DATA_DIR = "../data/gamedata/"
DATA_FILE = "dd_event_schemas.csv"
```

```python
data = pd.read_csv(DATA_DIR+DATA_FILE)
```

```python
genres = data.drop_duplicates(['genre'])
#genres = genres['genre']
```

```python
selectGenre = data[data['genre'] == 'No genre']
drop=selectGenre.drop_duplicates(['gameid','eventname'])
```

```python
for genre in genres['genre']:
    selectGenre = data[data['genre'] == genre]#Select records with a specified genre
    dataAfterDrop=selectGenre.drop_duplicates(['gameid','eventname'])#Drop duplicate data 
    count1 = dataAfterDrop['eventname'].value_counts()#(normalize=True) #Count frequency of event occurance
    #count2 = count1.drop_duplicates['eventname']
    count2 = dataAfterDrop.drop_duplicates(['eventname'])
    num = len(count2)#Number of events 
    print(genre)
    print('Number of events:','' , num)
    print(count1[:10])
    fig = plt.figure(figsize=(6,5))
    count1[:10].sort_values().plot(kind='barh',title=genre)
    #for a,b in zip(x,y):
    #    plt.text(a, b+0.05, '%.0f' % b, ha='center', va= 'bottom',fontsize=7)
    plt.show();
    print("")
    #header1=['eventname','score']
    #header2=DataFrame(header)
    #header2.to_csv(DATA_DIR+genre+'.csv',sep=',', header=False, index=False)  
    count1.to_csv(DATA_DIR+genre+'.csv', header=False)
    count1.to_csv(DATA_DIR+'/genreScore/'+genre+'Score.csv', header=False)
```
